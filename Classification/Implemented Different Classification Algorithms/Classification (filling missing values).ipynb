```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("breast cancer.csv")
```

```python
df.head()
```

```python
print (df.isnull().sum())
```

```python
df[['Clump','Marginal Adhesion']]
```

```python
print (df[['Clump','Marginal Adhesion']].isnull())
```

```python
freq_mean = df['Clump'].mean()
print (freq_mean)
df['Clump'].fillna(freq_mean, inplace=True)
```

```python
chord_mean = df['Marginal Adhesion'].mean()
print (chord_mean)
df['Marginal Adhesion'].fillna(chord_mean, inplace=True)
```
