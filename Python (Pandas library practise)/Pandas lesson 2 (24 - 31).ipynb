```python
#30
# Reindex is a method of DataFrame,iske ander aik parameter hai jska naam bhi method hi hai
# ffil aik value hai

#df bnaya, range and reshape use kr k
states = pd.DataFrame(np.arange(9).reshape (3,3), index = ['a','b','c'], columns = ['pehla', 'dusra', 'teesra'])
print (states, '\n')


#reindexing kr rhy hain ab
#added new column, jsmn NaN value show hrhi hai
##using reindex parameter (name is method), ffill = last column ki values uthakr place krdeta hai
# row per apply hga
# alphabetical and numbers order matter for method ffill and bfill ( random strings per error dega)
states = states.reindex(['b','c','a','d'], method = 'ffill')
print (states, '\n')

#adding column using reindex, ffill apply nhn hga idhr
states = states.reindex(columns = ['pehla', 'teesra', 'dusra', 'chouta'])
print (states, '\n')
```

```python
#29 Reindex Method

#reindex with range
a = pd.DataFrame(['talha','ali','hamza'], index = [0,3,6]) 
print (a, '\n')

#rows barhgyn, values usi index per rhi jisper assign kri thi
#range() python mn use hua krta tha
b = a.reindex(range(9))
print (b, '\n')

#filling values
c = a.reindex(range(9), method = 'ffill')
print (c, '\n')

#backfill bfll()
d = a.reindex(range(2,8), method = 'bfill')
print (d)
```

```python
#27 Deleting Data

data_df = pd.DataFrame(np.arange(16).reshape(4, 4), 
             index=['one', 'two', 'three', 'four'],
             columns=['ohio', 'colorado', 'utah', 'new York'])
print(data_df, '\n')

#deleting row, automatically axis = 0 hai (which means row)
# jo hmne naam define kra hai index k wohi mention krskty hain buss (inital index sey nhn krskty)
data_df = data_df.drop('two')
print(data_df, '\n')

#deleting row,  axis = 1  (which means columns), mention krna zarori hai
data_df = data_df.drop('colorado', axis =1)
print(data_df, '\n')

#inplace = true meaning - main dataframe per effect hu, ksi variable mn save na krwarhy hun tu
data_df.drop('four', inplace = True)
print(data_df, '\n')
```

```python
# Indexing, Selection, and Filtering

data_df = pd.DataFrame(np.arange(16).reshape((4, 4)), 
             index=['Ohio', 'Colorado', 'Utah', 'New York'],
             columns=['one', 'two', 'three', 'four'])
print(data_df, "\n")

#indexing is a subset of selection
# Selection (using column names) - printing columns
# ander brackets [] zarori hain
df2 = data_df[ ["one", "three"] ]
print(df2, "\n")

# filtering (using index number) - printing rows
df3 = data_df[2:] 
print(df3, "\n")

# dicitonary like style of accessing data
print(data_df["one"] , "\n")

# dictionary style k ander column k ander ki rows per apply hga
print(data_df["one"] [1:3], "\n")

# attribute like style of accessing data
print (data_df.one, "\n")

# filter on both row and column, column 'one' k 0 and 1st index value show hgi
print(data_df.one[0:2], "\n") 

# Conditonal Filtering/Selection
# column 3 ki woh values show krega which are greater then 5
print ( data_df[data_df['three'] > 5] , "\n")

# Boolean filtering/conditioning
f2 = data_df["three"] > 5
print(f2, "\n")

# sirf woh wali rows show krega jdhr column 'three' is greater then 5
print( data_df[  data_df["three"] > 5  ] )
```

```python
#31 Selection with Loc and iloc
# loc mn column, index names k through slicing hti hai
# iloc mn with index, column predifined numbers


data_df = pd.DataFrame(np.arange(16).reshape((4, 4)), 
             index=['Ohio', 'Colorado', 'Utah', 'New York'],
             columns=['one', 'two', 'three', 'four'])
print(data_df, "\n")

# in the loc method specify row label first, then specify column names
# remember! mutiple column names require array notation
# sbhse pehle row names (user defined), then hmare define krewe column names
print( data_df.loc[['Colorado','Ohio'], ['two',  'three'] ], "\n")

#other direct method
# sbhse pehle column name (user defined), then IMPORTANT index number (computer defined)
print(data_df[['two',  'three']] [1:3], "\n")

# iloc - using number instead of labels/names
# 2: (3rd and 4th row), [3, 0, 1] (4th then 1st then 2nd column)
print( data_df.iloc[2:, [3, 0, 1] ] , "\n")
print ( data_df.iloc[ :3 , :3 ] , "\n")
```

```python
#25 Arithmetic & Data Alignment

df1 = pd.DataFrame(np.arange(9).reshape((3, 3)), 
                   columns=list('bcd'), 
                   index=['Ohio', 'Texas', 'Colorado'])

df2 = pd.DataFrame(np.arange(12).reshape((4, 3)), 
                   columns=list('bde'),
                    index=['Utah', 'Ohio', 'Texas', 'Oregon'])

print(df1, '\n')
print(df2, '\n')
print()

# applying plus operation between two data frames
# NOTE: jin rows k naam same hai udhr addition hui hai bus, 
# NOTE: jin columns k naam same hai buss udhr addition hui hai
df3 = df1 + df2
print(df3, '\n')
```

```python
#26 Arithmetic Methods with Fill Values

df1 = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20).reshape((4, 5)),
                    columns=list('abcde'))
print(df1, '\n')
print(df2, '\n')

# using loc method to replace value, 1 row/ b column
df2.loc[1, 'b'] = 99
print(df2)

# direct + operation without fill_value
df3 = df1 + df2
print(df3, '\n')

# We can use add method for filling NaN cells with a value
# Nan will be replaced by 0 and then addition operation will apply
# ddition using a method with replacing Nan with 0,1,2,3,4,5
df3 = df1.add(df2, fill_value=1000)
print(df3)
```
