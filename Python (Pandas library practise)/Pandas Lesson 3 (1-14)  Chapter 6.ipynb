```python
# 3. Loading data from csv file

# simple loading file
# jo first row hai csv file ki automatically column names cinsider hgaya
sample = pd.read_csv('examples/sample.csv')
display (sample)

# assigning column names
# DataFrame mn columns parameter use hta hai, idhr names use hrha hai
# issue - 1st row ko bhi data smjh lea, for that we use skiprows = 1
sample2 = pd.read_csv('examples/sample.csv', skiprows = 1,  names = ['a', 'b', 'c', 'd', 'e', 'f'])
display (sample2)

# if 1st row ko column labels consider nhn krne and hum khud bhi assign nhn kr rhy naam
sample3 = pd.read_csv('examples/sample.csv', header = None)
display (sample3)
```

```python
# index names
sample = pd.read_csv ('examples/sample2.csv')
display (sample)

# first 2 cols as an index show hrhry hain ab
sample2 = pd.read_csv ('examples/sample2.csv', index_col = ['key1', 'key2'])
display (sample2)
```

```python
# 4. Handling missing values

# data mn different tareqe sey missing values hskti hai 
# examples NaN, NA,, nan, empty hu, not available, hmn identify krna hga
sample = pd.read_csv('examples/sample3.csv')
display (sample)

# sari nan values ko aik naam drhy hain takey easy hjae read krna
# aik dictionary bnaengy jsmn sari possible nan values jo hskti hain woh rkhngy
# list inside a dict
empty_values = {'message': ['foo', 'NA']}
sample2 = pd.read_csv('examples/sample3.csv', na_values=empty_values)
sample2
```

# Handling missing values is an important and frequently nuanced part of the file parsing process. Missing data is usually either not present (empty string) or marked by some sentinel value. By default, pandas uses a set of commonly occurring sentinels, such as NA and NULL

```python
# 5. Reading and text files

# apni marzi ki nmber of rows read krwarhy hain, data bht bara hai
sample = pd.read_csv('dataset/train.csv', nrows= 100)
display (sample)

```

```python
# reading a large dataset files in small chunks
# direct print nhn krwaskty/ loop lgana hga
sample2 = pd.read_csv('dataset/train.csv', chunksize = 500)
print (sample2)

chunklist = []
for chunk in sample2:
    chunklist.append(chunk)
    print (len(chunk), type(chunklist))
    
df = chunklist[0] 
display(df.head())

df2 = chunklist[3]
display(df2.head())
```

```python
# 6. Reading from multispace

# txt file sey data read krna hai, delimeter ismn space hai not comma
data = list(open('examples/ex3.txt'))
print (data)
display (data)

# sep parameter hai, seperator. s+ = use spaces (one or more) as a delimeter
sample = pd.read_table('examples/ex3.txt', sep='\s+')
display (sample)

# how to skip rows
sample2 = pd.read_csv('examples/ex4.csv', skiprows = [0,2,3])
display (sample2)

# table 6.2 of python for data analysis contains read_csv/read_table function arguements
```

```python
# 7. writing data to text format

# first reading data
sample = pd.read_csv('examples/ex5.csv')

# writing data, save the data into new file
sample.to_csv('examples/talha5.csv')

#now reading the file we have saved
sample2 = pd.read_csv('examples/talha5.csv')
display (sample2)

# date ki rannge bnae, new function used
sample3 = pd.date_range('1/1/2000', periods = 7)
display (sample3)

# then numpy ki range bnae and date (range) ko iska index bnadea
ts = pd.Series (np.arange(7), index = sample3)
display (ts)

# now saving it in new file 
ts.to_csv ("examples/wahh.csv")
```
